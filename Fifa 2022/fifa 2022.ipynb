## Fifa 2022 Web Scraping

```python
import numpy as np
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup as Soup
```

```python
url = "https://sofifa.com/players?offset="
```

```python
column = ['ID','Name','Full_Name','picture','Flag','Age','Position','Overall','Potential',
          'Team_Image','Team','Value', 'Nationality']
```

```python
fifa2022_dataframe = pd.DataFrame(columns = column)
```

```python
fifa2022_dataframe
```

```python
for offset in range(0, 301): # 301 -> https://sofifa.com/players?offset=18000
    url = url + str(offset*60)
    print(url)
    p_html = requests.get(url)
    p_soup = p_html.text
    data = Soup(p_soup,'html.parser')
    table = data.find('tbody')
    for i in table.findAll('tr'):    
        td = i.findAll('td')
        picture = td[0].find('img').get('data-src')
        ID = td[0].find('img').get('id')
        flag = td[1].find('img').get('data-src')
        Nationality = td[1].find('img').get('title')
        Name = td[1].findAll('a')[0].text
        Full_Name = td[1].findAll('a')[0]['data-tooltip']
        Age = td[2].text
        Position = td[1].findAll('a')[1].text
        Overall = td[3].find('span').text
        Potential = td[4].find('span').text
        Team_image = td[5].find('img').get('data-src')
        Team = td[5].find('a').text
        Value = td[6].text.strip()
        player_data = pd.DataFrame([[ID,Name,Full_Name,picture,flag,Age,Position,Overall,Potential,
                                 Team_image,Team,Value,Nationality]])
        #print(player_data)
        player_data.columns = column
        fifa2022_dataframe = fifa2022_dataframe.append(player_data, ignore_index = True)
    url = url.replace(str(offset*60), '')  
```

```python
fifa2022_dataframe.drop_duplicates(inplace = True)
```

```python
# Get detailed player information from player page
detailed_columns = ['Preferred Foot', 'Best Position', 'Jersey Number', 'Height', 'Weight','Crossing', 'Finishing', 'Dribbling', 'Long Passing', 'Ball Control', 'Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance', 'Shot Power', 'Jumping', 'Stamina', 'Strength', 'Aggression','Positioning', 'Defensive Awareness','ID']
detailed_data = pd.DataFrame(index = range(0, fifa2022_dataframe.count()[0]), columns = detailed_columns)
detailed_data.ID = fifa2022_dataframe.ID.values
```

```python
player_data_url = 'https://sofifa.com/player/'
for id in fifa2022_dataframe.ID:
    print(id)
    url = player_data_url + str(id)
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = Soup(plain_text, 'html.parser')
    skill_map = {}
    columns = soup.findAll('div', {'class': 'column col-4'})
    for column in columns:
        skills = column.findAll('li')
        for skill in skills:
            if(skill.find('label') != None):
                label = skill.find('label').text
                value = skill.text.replace(label, '').strip()
                skill_map[label] = value
    
    meta_data = soup.find('div', {'class': 'meta'}).text.split(' ')
    length = len(meta_data)
    weight = meta_data[length - 1]
    #height = meta_data[length - 2].split('\'')[0] + '\'' + meta_data[length - 2].split('\'')[1]
    height = meta_data[length - 2]
    skill_map["Height"] = height
    skill_map['Weight'] = weight
    features = soup.findAll('ul', {'class': 'pl'})

    try:
        preferred_foot_label = features[0].find('label').text
        preferred_foot_all_text = features[0].find('li', {'class': 'bp3-text-overflow-ellipsis'}).text
        preferred_foot_value = preferred_foot_all_text.replace(preferred_foot_label, '').strip()
        skill_map[preferred_foot_label] = preferred_foot_value
    except:
        skill_map['Preferred Foot'] = ""

    try:    
        jersey_number_label = features[2].findAll('label')[1].text
        jersey_number_all_text = features[2].findAll('li')[2].text
        jersey_number_value = jersey_number_all_text.replace(jersey_number_label, '').strip()
        skill_map[jersey_number_label] = jersey_number_value
    except:
        skill_map['Jersey Number'] = ""

    try:
        crossing_all_text = features[3].findAll('li')[0].text.strip()
        if("★" in crossing_all_text or "☆" in crossing_all_text):
            crossing_all_text = features[4].findAll('li')[0].text.strip()
            
        # print(crossing_all_text)
        if('Crossing' in crossing_all_text):
            crossing_value = crossing_all_text.split(" ")[0]
            if("+" in crossing_value):
                crossing_value = int(crossing_value.split("+")[0]) + int(crossing_value.split("+")[1])
            if("-" in crossing_value):
                crossing_value = int(crossing_value.split("-")[0]) - int(crossing_value.split("-")[1])
            crossing_label = crossing_all_text.split(" ")[1]
            skill_map[crossing_label] = crossing_value
        else:
            skill_map['Crossing'] = ""
    except:
        skill_map['Crossing'] = ""

    try:        
        finishing_all_text = features[3].findAll('li')[1].text.strip()
        if(not 'Finishing' in finishing_all_text):
               finishing_all_text = features[4].findAll('li')[1].text.strip()
                
        # print(finishing_all_text)        
        if('Finishing' in finishing_all_text):        
            finishing_value = finishing_all_text.split(" ")[0]
            if("+" in finishing_value):
                finishing_value = int(finishing_value.split("+")[0]) + int(finishing_value.split("+")[1])
            if("-" in finishing_value):
                finishing_value = int(finishing_value.split("-")[0]) - int(finishing_value.split("-")[1])
            finishing_label = finishing_all_text.split(" ")[1]
            skill_map[finishing_label] = finishing_value
        else:
            skill_map['Finishing'] = ""
    except:
        skill_map['Finishing'] = ""

    try:        
        dribbling_all_text = features[4].findAll('li')[0].text.strip()
        if(not 'Dribbling' in dribbling_all_text):
               dribbling_all_text = features[5].findAll('li')[0].text.strip()
                
        # print(dribbling_all_text)        
        if('Dribbling' in dribbling_all_text):        
            dribbling_value = dribbling_all_text.split(" ")[0]
            if("+" in dribbling_value):
                dribbling_value = int(dribbling_value.split("+")[0]) + int(dribbling_value.split("+")[1])
            if("-" in dribbling_value):
                dribbling_value = int(dribbling_value.split("-")[0]) - int(dribbling_value.split("-")[1]) 
            dribbling_label = dribbling_all_text.split(" ")[1]
            skill_map[dribbling_label] = dribbling_value
        else:
            skill_map['Dribbling'] = ""
    except:
        skill_map['Dribbling'] = ""

    try:        
        long_passing_all_text = features[4].findAll('li')[3].text.strip()
        if(not 'Long Passing' in long_passing_all_text):
               long_passing_all_text = features[5].findAll('li')[3].text.strip()
        
        # print(long_passing_all_text) 
        if('Long Passing' in long_passing_all_text):         
            long_passing_value = long_passing_all_text.split(" ")[0]
            if("+" in long_passing_value):
                long_passing_value = int(long_passing_value.split("+")[0]) + int(long_passing_value.split("+")[1])
            if("-" in long_passing_value):
                long_passing_value = int(long_passing_value.split("-")[0]) - int(long_passing_value.split("-")[1])    
            long_passing_label = long_passing_all_text.split(" ")[1] + " " + long_passing_all_text.split(" ")[2]
            skill_map[long_passing_label] = long_passing_value
        else:
            skill_map['Long Passing'] = ""
    except:
        skill_map['Long Passing'] = ""

    try:     
        ball_control_all_text = features[4].findAll('li')[4].text.strip()
        if(not 'Ball Control' in ball_control_all_text):
               ball_control_all_text = features[5].findAll('li')[4].text.strip()
        
        # print(ball_control_all_text) 
        if('Ball Control' in ball_control_all_text):        
            ball_control_value = ball_control_all_text.split(" ")[0]
            if("+" in ball_control_value):
                ball_control_value = int(ball_control_value.split("+")[0]) + int(ball_control_value.split("+")[1])
            if("-" in ball_control_value):
                ball_control_value = int(ball_control_value.split("-")[0]) - int(ball_control_value.split("-")[1])
            ball_control_label = ball_control_all_text.split(" ")[1] + " " + ball_control_all_text.split(" ")[2]
            skill_map[ball_control_label] = ball_control_value
        else:
            skill_map['Ball Control'] = ""
    except:
        skill_map['Ball Control'] = ""

    try:    
        acceleration_all_text = features[5].findAll('li')[0].text.strip()
        if(not 'Acceleration' in acceleration_all_text):
               acceleration_all_text = features[6].findAll('li')[0].text.strip()
                
        # print(acceleration_all_text)        
        if('Acceleration' in acceleration_all_text):        
            acceleration_value = acceleration_all_text.split(" ")[0]
            if("+" in acceleration_value):
                acceleration_value = int(acceleration_value.split("+")[0]) + int(acceleration_value.split("+")[1])
            if("-" in acceleration_value):
                acceleration_value = int(acceleration_value.split("-")[0]) - int(acceleration_value.split("-")[1])
            acceleration_label = acceleration_all_text.split(" ")[1]
            skill_map[acceleration_label] = acceleration_value
        else:
            skill_map['Acceleration'] = ""
    except:
        skill_map['Acceleration'] = ""

    try:    
        sprint_speed_all_text = features[5].findAll('li')[1].text.strip()
        if(not 'Sprint Speed' in sprint_speed_all_text):
               sprint_speed_all_text = features[6].findAll('li')[1].text.strip()
                
        # print(sprint_speed_all_text)        
        if('Sprint Speed' in sprint_speed_all_text):         
            sprint_speed_value = sprint_speed_all_text.split(" ")[0]
            if("+" in sprint_speed_value):
                sprint_speed_value = int(sprint_speed_value.split("+")[0]) + int(sprint_speed_value.split("+")[1])
            if("-" in sprint_speed_value):
                sprint_speed_value = int(sprint_speed_value.split("-")[0]) - int(sprint_speed_value.split("-")[1])     
            sprint_speed_label = sprint_speed_all_text.split(" ")[1] + " " + sprint_speed_all_text.split(" ")[2]
            skill_map[sprint_speed_label] = sprint_speed_value
        else:
            skill_map['Sprint Speed'] = ""
    except:
        skill_map['Sprint Speed'] = ""

    try:    
        agility_all_text = features[5].findAll('li')[2].text.strip()
        if(not 'Agility' in agility_all_text):
               agility_all_text = features[6].findAll('li')[2].text.strip()
        
        # print(agility_all_text)
        if('Agility' in agility_all_text):        
            agility_value = agility_all_text.split(" ")[0]
            if("+" in agility_value):
                agility_value = int(agility_value.split("+")[0]) + int(agility_value.split("+")[1])
            if("-" in agility_value):
                agility_value = int(agility_value.split("-")[0]) - int(agility_value.split("-")[1])          
            agility_label = agility_all_text.split(" ")[1]
            skill_map[agility_label] = agility_value
        else:
            skill_map['Agility'] = ""
    except:
        skill_map['Agility'] = ""

    try:    
        reactions_all_text = features[5].findAll('li')[3].text.strip()
        if(not 'Reactions' in reactions_all_text):
               reactions_all_text = features[6].findAll('li')[3].text.strip()
        
        # print(reactions_all_text)
        if('Reactions' in reactions_all_text):        
            reactions_value = reactions_all_text.split(" ")[0]
            if("+" in reactions_value):
                reactions_value = int(reactions_value.split("+")[0]) + int(reactions_value.split("+")[1])
            if("-" in reactions_value):
                reactions_value = int(reactions_value.split("-")[0]) - int(reactions_value.split("-")[1])        
            reactions_label = reactions_all_text.split(" ")[1]
            skill_map[reactions_label] = reactions_value
        else:
            skill_map['Reactions'] = ""
    except:
        skill_map['Reactions'] = ""

    try:    
        balance_all_text = features[5].findAll('li')[4].text.strip()
        if(not 'Balance' in balance_all_text):
               balance_all_text = features[6].findAll('li')[4].text.strip()
        
        # print(balance_all_text)
        if('Balance' in balance_all_text):         
            balance_value = balance_all_text.split(" ")[0]
            if("+" in balance_value):
                balance_value = int(balance_value.split("+")[0]) + int(balance_value.split("+")[1])
            if("-" in balance_value):
                balance_value = int(balance_value.split("-")[0]) - int(balance_value.split("-")[1])        
            balance_label = balance_all_text.split(" ")[1]
            skill_map[balance_label] = balance_value
        else:
            skill_map['Balance'] = ""
    except:
        skill_map['Balance'] = ""

    try:    
        shot_power_text = features[6].findAll('li')[0].text.strip()
        if(not 'Shot Power' in shot_power_text):
               shot_power_text = features[7].findAll('li')[0].text.strip()
        
        # print(shot_power_text)
        if('Shot Power' in shot_power_text):        
            shot_power_value = shot_power_text.split(" ")[0]
            if("+" in shot_power_value):
                shot_power_value = int(shot_power_value.split("+")[0]) + int(shot_power_value.split("+")[1])
            if("-" in shot_power_value):
                shot_power_value = int(shot_power_value.split("-")[0]) - int(shot_power_value.split("-")[1])         
            shot_power_label = shot_power_text.split(" ")[1] + " " + shot_power_text.split(" ")[2]
            skill_map[shot_power_label] = shot_power_value
        else:
            skill_map['Shot Power'] = ""
    except:
        skill_map['Shot Power'] = ""

    try:    
        jumping_text = features[6].findAll('li')[1].text.strip()
        if(not 'Jumping' in jumping_text):
               jumping_text = features[7].findAll('li')[1].text.strip()
        
        # print(jumping_text)
        if('Jumping' in jumping_text):        
            jumping_value = jumping_text.split(" ")[0]        
            if("+" in jumping_value):
                jumping_value = int(jumping_value.split("+")[0]) + int(jumping_value.split("+")[1])
            if("-" in jumping_value):
                jumping_value = int(jumping_value.split("-")[0]) - int(jumping_value.split("-")[1])        
            jumping_label = jumping_text.split(" ")[1]
            skill_map[jumping_label] = jumping_value
        else:
            skill_map['Jumping'] = ""
    except:
        skill_map['Jumping'] = ""

    try:    
        stamina_text = features[6].findAll('li')[2].text.strip()
        if(not 'Stamina' in stamina_text):
               stamina_text = features[7].findAll('li')[2].text.strip()
        
        # print(stamina_text)
        if('Stamina' in stamina_text):        
            stamina_value = stamina_text.split(" ")[0]
            if("+" in stamina_value):
                stamina_value = int(stamina_value.split("+")[0]) + int(stamina_value.split("+")[1])
            if("-" in stamina_value):
                stamina_value = int(stamina_value.split("-")[0]) - int(stamina_value.split("-")[1])        
            stamina_label = stamina_text.split(" ")[1]
            skill_map[stamina_label] = stamina_value
        else:
            skill_map['Stamina'] = ""
    except:
        skill_map['Stamina'] = ""

    try:    
        strength_text = features[6].findAll('li')[3].text.strip()
        if(not 'Strength' in strength_text):
               strength_text = features[7].findAll('li')[3].text.strip()
        
        # print(strength_text)
        if('Strength' in strength_text):         
            strength_value = strength_text.split(" ")[0]
            if("+" in strength_value):
                strength_value = int(strength_value.split("+")[0]) + int(strength_value.split("+")[1])
            if("-" in strength_value):
                strength_value = int(strength_value.split("-")[0]) - int(strength_value.split("-")[1])            
            strength_label = strength_text.split(" ")[1]
            skill_map[strength_label] = strength_value
        else:
            skill_map['Strength'] = ""
    except:
        skill_map['Strength'] = ""

    try:    
        aggression_text = features[7].findAll('li')[0].text.strip()
        if(not 'Aggression' in aggression_text):
               aggression_text = features[8].findAll('li')[0].text.strip()
        
        # print(aggression_text)
        if('Aggression' in aggression_text):         
            aggression_value = aggression_text.split(" ")[0]
            if("+" in aggression_value):
                aggression_value = int(aggression_value.split("+")[0]) + int(aggression_value.split("+")[1])
            if("-" in aggression_value):
                aggression_value = int(aggression_value.split("-")[0]) - int(aggression_value.split("-")[1])
            aggression_label = aggression_text.split(" ")[1]
            skill_map[aggression_label] = aggression_value
        else:
            skill_map['Aggression'] = ""
    except:
        skill_map['Aggression'] = ""

    try:    
        positioning_text = features[7].findAll('li')[2].text.strip()
        if(not 'Positioning' in positioning_text):
               positioning_text = features[8].findAll('li')[2].text.strip()
        
        # print(positioning_text)
        if('Positioning' in positioning_text):         
            positioning_value = positioning_text.split(" ")[0]
            if("+" in positioning_value):
                positioning_value = int(positioning_value.split("+")[0]) + int(positioning_value.split("+")[1])
            if("-" in positioning_value):
                positioning_value = int(positioning_value.split("-")[0]) - int(positioning_value.split("-")[1])
            positioning_label = positioning_text.split(" ")[1]
            skill_map[positioning_label] = positioning_value
        else:
            skill_map['Positioning'] = ""
    except:
        skill_map['Positioning'] = ""

    try:    
        defensive_awareness_text = features[8].findAll('li')[0].text.strip()
        if(not 'Defensive Awareness' in defensive_awareness_text):
               defensive_awareness_text = features[9].findAll('li')[0].text.strip()
        
        # print(defensive_awareness_text)
        if('Defensive Awareness' in defensive_awareness_text):        
            defensive_awareness_value = defensive_awareness_text.split(" ")[0]
            if("+" in defensive_awareness_value):
                defensive_awareness_value = int(defensive_awareness_value.split("+")[0]) + int(defensive_awareness_value.split("+")[1])
            if("-" in defensive_awareness_value):
                defensive_awareness_value = int(defensive_awareness_value.split("-")[0]) - int(defensive_awareness_value.split("-")[1])
            defensive_awareness_label = defensive_awareness_text.split(" ")[1] + " " + defensive_awareness_text.split(" ")[2]
            skill_map[defensive_awareness_label] = defensive_awareness_value
        else:
            skill_map['Defensive Awareness'] = ""
    except:
        skill_map['Defensive Awareness'] = ""
        
    # print(skill_map)    
        
    for key, value in skill_map.items():
        detailed_data.loc[detailed_data.ID == id, key] = value
```
