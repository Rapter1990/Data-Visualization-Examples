# Web Scraping for Country Name and Country Images

```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import requests 
```

```python
BASE_URL = "http://www.countryflags.org/"
```

```python
#loading empty array for board members
countries = []
```

```python
html = requests.get(BASE_URL).text
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
#identify table we want to scrape
officer_table = soup.find('table')
#print(officer_table)
```

```python
try:
   #loop through table, grab each of the 4 columns shown (try one of the links yourself to see the layout)
   for row in officer_table.find_all('tr'):
        cols = row.find_all('td')
        #print(cols)
        for img in row.find_all('img'):
            #print(img)
            country_name = img['alt']
            country_image = BASE_URL + img['src']
            country_image = country_image.replace("!@#$%^&*()'[]{};:,./<>?\|`~-=_+", " ")
            countries.append((country_name,country_image))
except: pass  
```

```python
countries
```

```python
country_df = pd.DataFrame(countries)
```

```python
country_df.columns = ['Country Name', 'Image URL']
```

```python
country_df = country_df.reset_index()
```
