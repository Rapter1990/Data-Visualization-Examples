16 - Define meltDataFrame to determine Date and Total_Ratio for each IDE during time period

```python
def meltDataFrame(df,id_vars,value_vars,var_name,value_name):
    return pd.melt(df,
        id_vars= id_vars,
        value_vars = value_vars,
        var_name= var_name,
        value_name= value_name)


popular_ide_df_melt = meltDataFrame(popular_ide_df,
        ["index"],popular_ide_df.columns[1:],'Date','Total_Ratio')
```

```python
popular_ide_df_melt
```

17 - Change "index" name to "IDE"

```python
popular_ide_df_melt.rename(columns={'index':'IDE'}, inplace=True)
```

18 - Select Date from 2004 to 2012 and from 2012 to 2020

```python
popular_ide_df_melt_2004_2012 = popular_ide_df_melt[ \
                                                    popular_ide_df_melt['Date'].str.contains("2004") | \
                                                    popular_ide_df_melt['Date'].str.contains("2005") | \
                                                    popular_ide_df_melt['Date'].str.contains("2006") | \
                                                    popular_ide_df_melt['Date'].str.contains("2007") | \
                                                    popular_ide_df_melt['Date'].str.contains("2008") | \
                                                    popular_ide_df_melt['Date'].str.contains("2009") | \
                                                    popular_ide_df_melt['Date'].str.contains("2010") | \
                                                    popular_ide_df_melt['Date'].str.contains("2011") | \
                                                    popular_ide_df_melt['Date'].str.contains("2012") \
                                                   ] 

popular_ide_df_melt_2012_2020 = popular_ide_df_melt[ \
                                                    popular_ide_df_melt['Date'].str.contains("2013") | \
                                                    popular_ide_df_melt['Date'].str.contains("2014") | \
                                                    popular_ide_df_melt['Date'].str.contains("2015") | \
                                                    popular_ide_df_melt['Date'].str.contains("2016") | \
                                                    popular_ide_df_melt['Date'].str.contains("2017") | \
                                                    popular_ide_df_melt['Date'].str.contains("2018") | \
                                                    popular_ide_df_melt['Date'].str.contains("2019") | \
                                                    popular_ide_df_melt['Date'].str.contains("2020") \
                                                   ] 
```

17 - Draw a Timeline to show the popularity of IDE Since 2004

```python
custom_style = {'axes.labelcolor': 'black',
                'axes.edgecolor': 'black',
                'xtick.color': 'black',
                'ytick.color': 'black',
                'legend.frameon':True
               }

sns.set_style("darkgrid", rc=custom_style)
```

```python
plt.figure(figsize=(18,9))

ax = sns.pointplot(data = popular_ide_df_melt_2004_2012, 
                   x='Date', 
                   y='Total_Ratio', 
                   hue='IDE', 
                   palette='rainbow',
                   facet_kws={"legend_out": True}                   
                  )
plt.title('Timeline for Most Popular IDE from 2004-2012', fontsize=22, y=1.015)
plt.xlabel('date(month-year)', labelpad=16)
plt.ylabel('percentage', labelpad=16)
leg = ax.legend(loc='center right', bbox_to_anchor=(1.15, 0.5), ncol=1)
plt.xticks(rotation=90)
plt.setp(ax.collections, alpha=.6) #for the markers
plt.setp(ax.lines, alpha=.3) #for the lines
plt.savefig('images/image1.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(18,9))

ax = sns.pointplot(data = popular_ide_df_melt_2012_2020, 
                   x='Date', 
                   y='Total_Ratio', 
                   hue='IDE', 
                   palette='rainbow',
                   facet_kws={"legend_out": True}
                  )
plt.title('Timeline for Most Popular IDE from 2012-2020', fontsize=22, y=1.015)
plt.xlabel('date(month-year)', labelpad=16)
plt.ylabel('percentage', labelpad=16)
leg = ax.legend(loc='center right', bbox_to_anchor=(1.15, 0.5), ncol=1)
plt.xticks(rotation=90)
plt.setp(ax.collections, alpha=.6) #for the markers
plt.setp(ax.lines, alpha=.3) #for the lines
plt.savefig('images/image2.png', bbox_inches='tight')
plt.show()
```

18 - WorldCloud showing the names of IDE

```python
plt.figure(figsize=[15, 15])

image = Image.open("images/ide.png")
image_gray = image.convert('L')
image_invert = ImageOps.invert(image_gray)
char_mask = np.array(image_invert)
image_colors = ImageColorGenerator(char_mask)


wordcloud = WordCloud(stopwords=STOPWORDS,
                        background_color='black',
                        #width=1000,  
                        #height=1000, 
                        max_words=300,
                        mask=char_mask,
                        contour_width=3, 
                        contour_color='white',
                    ).generate(' '.join(popular_ide_df_melt['IDE'].unique()))


plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Names of IDE', fontsize=20)
plt.savefig('images/image3.png')
plt.show()
```
